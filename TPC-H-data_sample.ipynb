```python
import pandas as pd
```

```python
#download data: https://docs.deistercloud.com/Databases.30/TPCH%20Benchmark.90/Data%20generation%20tool.30.xml?embedded=true
#CUSTOMER = pd.read_table('data/customer.csv', header=None, sep = '|')
LINEITEM = pd.read_table('data/lineitem.tbl', header=None, sep = '|')
NATION = pd.read_table('data/nation.tbl', header=None, sep='|')
ORDERS = pd.read_table('data/orders.tbl', header=None, sep = '|')
PART = pd.read_table('data/part.tbl', header=None, sep = '|')
PARTSUPP = pd.read_table('data/partsupp.tbl', header=None, sep = '|')
REGION = pd.read_table('data/region.tbl', header=None, sep = '|')
SUPPLIER = pd.read_table('data/supplier.tbl', header=None, sep = '|')
```

```python
#cleaning http://www.tpc.org/tpc_documents_current_versions/pdf/tpc-h_v2.18.0.pdf section 1.4
#remove column of NAs
CUSTOMER = CUSTOMER.iloc[:, :-1]
LINEITEM = LINEITEM.iloc[:, :-1]
NATION = NATION.iloc[:, :-1]
ORDERS = ORDERS.iloc[:, :-1]
PART = PART.iloc[:, :-1]
PARTSUPP = PARTSUPP.iloc[:, :-1]
REGION = REGION.iloc[:, :-1]
SUPPLIER = SUPPLIER.iloc[:, :-1]
```

```python
#assign column names
CUSTOMER.columns = ['CUSTKEY', 'NAME', 'ADDRESS', 'NATIONKEY', 'PHONE', 'ACCTBAL', 
                    'MKTSEGMENT', 'COMMENT']
LINEITEM.columns = ['ORDERKEY', 'PARTKEY', 'SUPPKEY', 'LINENUMBER', 'QUANTITY', 'EXTENDEDPRICE',
                    'DISCOUNT', 'TAX','RETURNFLAG', 'LINESTATUS', 'SHIPDATE', 'COMMITDATE',
                    'RECEIPTDATE', 'SHIPINSTRUCT', 'SHIPMODE', 'COMMENT']
NATION.columns = ['NATIONKEY', 'NAME', 'REGIONKEY', 'COMMENT']
ORDERS.columns = ['ORDERKEY', 'CUSTKEY', 'ORDERSTATUS', 'TOTALPRICE', 'ORDERDATE', 'ORDERPRIORITY',
                  'CLERK', 'SHIPPRIORITY', 'COMMENT']
PART.columns = ['PARTKEY', 'NAME','MFGR', 'BRAND', 'TYPE', 'SIZE', 'CONTAINER', 'RETAILPRICE', 
               'COMMENT']
PARTSUPP.columns = ['PARTKEY', 'SUPPKEY', 'AVAILQTY', 'SUPPLYCOST', 'COMMENT']
REGION.columns = ['REGIONKEY', 'NAME', 'COMMENT']
SUPPLIER.columns = ['SUPPKEY', 'NAME', 'ADDRESS', 'NATIONKEY', 'PHONE', 'ACCTBAL', 'COMMENT']
```

```python
#remove comment columns
CUSTOMER = CUSTOMER.drop('COMMENT', axis=1)
LINEITEM = LINEITEM.drop('COMMENT', axis=1)
NATION = NATION.drop('COMMENT', axis=1)
ORDERS = ORDERS.drop('COMMENT', axis=1)
PART = PART.drop('COMMENT', axis=1)
PARTSUPP = PARTSUPP.drop('COMMENT', axis=1)
REGION = REGION.drop('COMMENT', axis=1)
SUPPLIER = SUPPLIER.drop('COMMENT', axis=1)
```

```python
#export as csv
CUSTOMER.to_csv('data/customer_clean.csv')
LINEITEM.to_csv('data/lineitem_clean.csv')
NATION.to_csv('data/nation_clean.csv')
ORDERS.to_csv('data/orders_clean.csv')
PART.to_csv('data/part_clean.csv')
PARTSUPP.to_csv('data/partsupp_clean.csv')
REGION.to_csv('data/region_clean.csv')
SUPPLIER.to_csv('data/supplier_clean.csv')
```
